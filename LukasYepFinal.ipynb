```
class Country():
  def __init__(self, name):
    self.name = name

    #clean series and define all self.x values
  def populate(self):
    adfertseries = adfert[self.name].values
    adfertseries = adfertseries[~np.isnan(adfertseries)]
    if len(adfertseries) > 0:
      self.adolescentfertility = adfertseries[-1]
    else:
      self.adfertseries = 0
    birthrateseries = birthrate[self.name].values
    birthrateseries = birthrateseries[~np.isnan(birthrateseries)]
    if len(birthrateseries) > 0:
      self.birth_rate = birthrateseries[-1]
    else:
      self.birth_rate = 0
    gdpseries = gdp[self.name].values
    gdpseries = gdpseries[~np.isnan(gdpseries)]
    if len(gdpseries) > 0:
      self.total_gdp = gdpseries[-1]
    else:
      self.total_gdp = 0
    gdpgrowthseries = gdpgrowth[self.name].values
    gdpgrowthseries = gdpgrowthseries[~np.isnan(gdpgrowthseries)]
    if len(gdpgrowthseries) > 0:
      self.gdp_growth = gdpgrowthseries[-1]
    else:
      self.gdp_growth = 0
    gdppcseries = gdppc[self.name].values
    gdppcseries = gdppcseries[~np.isnan(gdppcseries)]
    if len(gdppcseries) > 0:
      self.gdppercapita = gdppcseries[-1]
    else:
      self.gdppercapita = 0
    gdppcgrowthseries = gdppcgrowth[self.name].values
    gdppcgrowthseries = gdppcgrowthseries[~np.isnan(gdppcgrowthseries)]
    if len(gdppcgrowthseries) > 0:
      self.gdppc_growth = gdppcgrowthseries[-1]
    else:
      self.gdppc_growth = 0
    gdpppersseries = gdpppers[self.name].values
    gdpppersseries = gdpppersseries[~np.isnan(gdpppersseries)]
    if len(gdpppersseries) > 0:
      self.gdpperperson = gdpppersseries[-1]
    else:
      self.gdpperperson = 0
    giniseries = gini[self.name].values
    giniseries = giniseries[~np.isnan(giniseries)]
    if len(giniseries) > 0:
      self.ginicoefficient = giniseries[-1]
    else:
      self.ginicoefficient = 0
    goldresseries = goldres[self.name].values
    goldresseries = goldresseries[~np.isnan(goldresseries)]
    if len(goldresseries) > 0:
      self.goldreserves = goldresseries[-1]
    else:
      self.goldreserves = 0
    infmortseries = infmort[self.name].values
    infmortseries = infmortseries[~np.isnan(infmortseries)]
    if len(infmortseries) > 0:
      self.infantmortality = infmortseries[-1]
    else:
      self.infantmortality = 0
    landsizeseries = landsize[self.name].values
    landsizeseries = landsizeseries[~np.isnan(landsizeseries)]
    if len(landsizeseries) > 0:
      self.land_size = landsizeseries[-1]
    else:
      self.land_size = 0
    matmortseries = matmort[self.name].values
    matmortseries = matmortseries[~np.isnan(matmortseries)]
    if len(matmortseries) > 0:
      self.maternalmortality = matmortseries[-1]
    else:
      self.maternalmortality = 0
    milspendseries = milspend[self.name].values
    milspendseries = milspendseries[~np.isnan(milspendseries)]
    if len(milspendseries) > 0:
      self.militaryspending = milspendseries[-1]
    else:
      self.militaryspending = 0
    rurpopseries = rurpop[self.name].values
    rurpopseries = rurpopseries[~np.isnan(rurpopseries)]
    if len(rurpopseries) > 0:
      self.ruralpop = rurpopseries[-1]
    else:
      self.ruralpop = 0
    u5mortseries = u5mort[self.name].values
    u5mortseries = u5mortseries[~np.isnan(u5mortseries)]
    self.under5mort = u5mortseries[-1]
    if self.ginicoefficient == 0:
      self.equalityscore = 'data is not available'
    if 0 < self.ginicoefficient <= 30:
      self.equalityscore = 'very equal'
    if 30 < self.ginicoefficient <= 37:
      self.equalityscore = 'fairly equal'
    if 37 < self.ginicoefficient <= 44:
      self.equalityscore = 'ok'
    if 44 < self.ginicoefficient <= 51:
      self.equalityscore = 'fairly unequal'
    if self.ginicoefficient > 51:
      self.equalityscore = 'very unequal'
    
  #this uses the data  
  def stats(self):
    print(f"Some economic stats about {self.name}:")
    print(f"The GDP of {self.name} is approximately ${(self.total_gdp)/1000000000000} trillion")
    print(f"Here's a nice little chart of {self.name}'s GDP since 1960:")
    ax = plt.gca()
    gdp.plot(kind='line',x='Year',y=self.name, color='blue', ax=ax)
    plt.show()
    print(f"And, here's a chart of the GDP growth:")
    ax = plt.gca()
    gdpgrowth.plot(kind='line',x='Year',y=self.name, color='blue', ax=ax)
    plt.show()
    print(f"Last year, their GDP grew by {round(self.gdp_growth,2)}%")
    print(f"And GDP per capita is ${round(self.gdppercapita,0)}")
    print(f"The gini coefficient is {self.ginicoefficient}, which means that wealth distribution is {self.equalityscore}")
    print(f"{self.name} spends {round(self.militaryspending,2)}% of their budget on the military")
    print(f"In {self.name}, there are {round(self.birth_rate,3)} births for every 1,000 people")
    print(f"For every 100,000 births in {self.name}:")
    print(f"An estimated {round(self.maternalmortality,2)} mothers die during birth")
    print(f"{round((self.infantmortality)*100,2)} infants die")
    print(f"{round((self.under5mort)*100,2)} children under 5 die.")
    print(f"And, among women 15-19, for every 1000 women there are {round(self.adolescentfertility,3)} births")
    print(f"{self.name} is {round(self.land_size,0)} square kilometers")
    print(f"... or {round(self.land_size*0.3861,0)} square miles")
    print(f"And they keep ${round((self.goldreserves)/1000000000,1)} billion in reserve")
    print(f"Finally, here's how the GDP of {self.name} compares to some other countries")
    ax = plt.gca()
    if self.name != 'Japan' and self.name != 'China' and self.name != 'United States' \
    and self.name != 'Germany' and self.name != 'India' and self.name != 'France':
      gdp.plot(kind='line',x='Year',y=self.name, color='black', ax=ax)
    else:
      gdp.plot(kind='line',x='Year',y='France', color='black', ax=ax)
    gdp.plot(kind='line',x='Year',y='China', color='red', ax=ax)
    gdp.plot(kind='line',x='Year',y='India', color='orange', ax=ax)
    gdp.plot(kind='line',x='Year',y='United States', color='blue', ax=ax)
    gdp.plot(kind='line',x='Year',y='Germany', color='gold', ax=ax)
    gdp.plot(kind='line',x='Year',y='Japan', color='mediumorchid', ax=ax)
    plt.show()
    
    
# making use of the code that's been written
'''
c1name = input("Enter the name of a country: ")    
C1 = Country(c1name)
C1.populate()
C1.stats()
'''

Japan = Country("Japan")
Japan.populate()
Japan.stats()
```
